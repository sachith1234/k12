```python
### Create a blank white color image (512x512x3)
img = np.zeros((512,512,3),np.uint8)+255
plt.imshow(img)
plt.show()
```

```python
### Drwa a line on image
img = np.zeros((512,512,3),np.uint8)+255
p1 = (50,300)
p2 = (300,50)
cv2.line(img,p1,p2,(100,155,10),10)
plt.imshow(img)
plt.show()
```

```python
### Draw a rectangle on image
img = np.zeros((512,512,3),np.uint8)+255
p1 = (50,300)
p2 = (300,50)
cv2.rectangle(img,p1,p2,(100,155,10),5)
plt.imshow(img)
plt.show()
```

```python
### Draw a rectangle on image
img = np.zeros((512,512,3),np.uint8)+255
p1 = (50,300)
p2 = (300,50)
cv2.rectangle(img,p1,p2,(34,67,23),-1)
plt.imshow(img)
plt.show()
```

```python
### Drwa a circle on image
img = np.zeros((512,512,3),np.uint8)+255
center = (250,250)
radius = 200
cv2.circle(img,center,radius,(100,155,10),5)
cv2.circle(img,center,20,(255,0,0),-1)
plt.imshow(img)
plt.show()
```

```python
### Drwa a ellipse on image
img = np.zeros((512,512,3),np.uint8)+255
center = (250,250)
axes = (200,50)   ##(major,minor)
angle = 0
startangle = 0
endangle = 360
cv2.ellipse(img,center,axes,angle,startangle,endangle,(100,155,10),5)
plt.imshow(img)
plt.show()
```

```python
## Write a text on image
img = np.zeros((512,512,3),np.uint8)+255
text = "Umang"
origin = (100,300)
style = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
size = 2.5
cv2.putText(img,text,origin,style,size,(100,155,10),2)
plt.imshow(img)
plt.show()
```

```python
## Save a image
save_path = r"C:\Users\UMANG KEJRIWAL\Desktop\Open CV\myimage.jpg"
img = np.zeros((512,512,3),np.uint8)+255
text = "Umang"
origin = (100,300)
style = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
size = 2.5
cv2.putText(img,text,origin,style,size,(100,155,10),2)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.imwrite(save_path,img)
plt.imshow(img)
plt.show()
```
